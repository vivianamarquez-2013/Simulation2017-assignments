```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib as mpl
import seaborn as sns 
```

```python
with open('Data_Simulation.txt', 'r') as f:
    lines = f.readlines()

yaxis = []    
for line in lines:
    yaxis.append(float(line))
    
max_item = max(yaxis)
min_item = min(yaxis)
xaxis = list(range(0,len(yaxis)))

mpl.rc("figure", figsize=(16, 8))
sns.set_palette("muted") #husl
```

```python
plt.xlim(0,len(yaxis))
plt.ylim(min_item,max_item)
plt.plot(xaxis,yaxis)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Data Plot')
plt.show()
```
