### Spectral Test

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline 

plt.title("Spectral Test - 2D")
plt.plot(yaxis[:-1],yaxis[1:],"mo")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
plt.title("Spectral Test - 3D")
ax.scatter(yaxis[:-2],yaxis[1:-1],yaxis[2:],c="m",marker="o")
```

Since the distance in 2D and 3D is constant for the numbers, then the sequence passes the Spectral test. 

### $\chi^2$ Test

```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

fig,ax = plt.subplots(1,1)

linestyles = ['-']
deg_of_freedom = [len(yaxis)]
for df, ls in zip(deg_of_freedom, linestyles):
  ax.plot(yaxis, stats.chi2.pdf(yaxis, df), linestyle=ls, c="m")

plt.xlabel('$\chi^2$')
plt.ylabel(r'$f(\chi^2)$')
plt.title(r'$\chi^2\ \mathrm{Distribution}$')

plt.legend()
plt.show()
```

```python
from scipy.stats import chisquare
#expected = np.sum(yaxis)/len(yaxis)
#chisquare(yaxis,expected)
chisquare(yaxis)
```

$\chi^2$ = $16584.236618490446$ <br>
$p$-value = $1.0$ <br><br>
It also passes the $\chi^2$ test.

### Kolmogorov-Smirnov Test

```python
stats.kstest(yaxis, "norm")
```

```python
stats.kstest(yaxis, 'uniform')
```

<font color="blue"><b>Permutation test:</b></font><br><br>
A permutation test gives us a way to compute the sampling distribution for any test statistic. Notice that by randomly shuffling the labels on data points when performing significance tests we can make up as many data sets as we like. In these situations, the conditional distribution of the test statistic under the null hypothesis is completely determined by the fact that all relabelings of the data are equally likely.

When we permute the outcome values during the test, we see all of the possible alternative treatment assignments we could have had and where the mean-difference in our observed data falls relative to all of the differences we could have seen if the outcome was independent of treatment assignment. In case the data set is "too" large, we can do an “approximate permutation tests” by conducting a large number of resamples. That process should approximate the permutation distribution.

As a fun fact, permutation tests are widely used in genetics and genomics. 

```python
semilla = 23
x = [] #My random numbers
u = []
m = 479
a = 234
c = 478

x.append((a*semilla+c)%m)

for i in range(0, 10):
    u.append(x[i]/m)
    x.append((a*x[i]+c)%m)
```
